```python
df=pd.read_csv("stock_data.csv",header=0,na_values={
    'eps':['n.a.','not available'],
    'revenue':[-1,'n.a.','not available'],
    'price':[-1,'n.a.','not available'],
    'people':[-1,'n.a.','not available']
})
df
```

```python
df.to_csv("clean_stocks.csv",index=False)
```

```python
def convert_people_cell(cell):
    if cell=='n.a.':
        return 'john doe'
    return cell

def convert_revenue_cell(cell):
    if cell==-1:
        return 0
    return cell

df=pd.read_excel("stock_data.xlsx","Sheet1",converters={
    'people':convert_people_cell,
    'revenue':convert_revenue_cell
})
df
```

```python
df
```

```python
df.to_excel("new.xlsx",sheet_name='stocks',index=False,startrow=1,startcol=2)
```
